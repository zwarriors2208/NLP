### count percentage of punctions in it

```python
import string
tokens = word_tokenize(sents)
count = 0
for i in tokens:
    if i.endswith(tuple(string.punctuation)):
        count += 1
                  
```

```python
count / len(tokens) * 100
```

```python
count = [word for word in tokens if not word.isalpha() and not word.isdigit()]
```

### various tokenizer
**whitespace Tokenizer**
- the space which seprate the words are called white space
- new line, tab, and space' ' are considered whitespace

**space tokenizer**
- only space is considered

**Line Tokenizer**
- only newline is taken as delimeter/ separator

**Tab Tokenizer**
- only tab (\t) is taken as separator

**Tweet Tokenizer** 
- idea of smileys created by tweeter
- :/          ;[ :D ;o

**MWE (Multi Word Extension) Tokenizer**
- join two words by add_mwe
- then output would group them together
 
**Custom Tokenizer**

### Custom Tokenizer

```python
import re
import string
display(string.punctuation)
def custom_tokenizer(text):
    return re.split(r'[.,;?!\s]+',text)

text = 'This is some text with punctuation! > Let\'s tokenize it. Is it ok?'

tokens= custom_tokenizer(text)

print('Tokens: ')
for token in tokens:
    print(token)
```
