```python
url = urlopen("https://en.wikipedia.org/wiki/Tiger")
html= url.read()
soup = BeautifulSoup(html, 'html.parser')
data = soup.get_text(strip = False)
tokens = word_tokenize(data)
tokens1 = []
for i in tokens:
    if not i.endswith(tuple(string.punctuation)) and i.lower() not in stopwords.words('english') and len(i) > 2:
        tokens1.append(i.lower())
tokens2 = [ls.stem(token) for token in tokens1]
freq = FreqDist(tokens2)
freq.plot(20)
freq.most_common(5)
```
